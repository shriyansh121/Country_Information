# Importing necessary Libraries

```python
import wikipediaapi
import json
```

```python
# Initialize Wikipedia API
user_agent = "MyCountryInfoScript/1.0 (https://mywebsite.com; myemail@gmail.com)"
wiki = wikipediaapi.Wikipedia(language='en',user_agent = user_agent)
```

# Function to find summaries of all countries with respect to their continents

```python
def generate_country_description(country_name, word_count=300):
    page = wiki.page(country_name)
    if page.exists():
        summary = ' '.join(page.summary.split()[:word_count])
        return summary
    else:
        return f"No information available for {country_name}."
```

# Australia

```python
# List of countries
australia = ["Australia", "New Zealand", "Papua New Guinea", "Nauru", "Palau","Micronesia"
             "Solomon Islands", "Vanuatu", "Fiji", "Tuvalu", "Samoa",
             "Kiribati", "Marshall Islands"]
```

```python
australia_summaries = {}
for country in australia:
    australia_summaries[country] = generate_country_description(country)
```

```python
print(australia_summaries)
```

```python
with open('Australasia_summaries.json','w') as f:
    json.dump(australia_summaries,f,indent=4)
```
